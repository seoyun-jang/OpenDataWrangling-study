* 경도, 위도 -> 숫자타입으로 바꿔주기

```python
x = shop['경도']
shop['경도'] = shop['경도'].map(lambda x : x.replace('"','')).astype(float)

y = shop['위도']
shop['위도'] = shop['위도'].map(lambda x : x.replace('"','')).astype(float)
```

```python
#위도와 경도에 따른 시각화 (1000개)

(ggplot(shop[:1000])
 + aes(x = '경도', y = '위도')
 + geom_point()
 + theme(text = element_text(family = 'NanumGothic'))
)
```

```python
shop[:1000].plot.scatter(x='경도', y='위도', grid=True)
```

* 서울과 부산으로 데이터 나누기

```python
shop_seoul = shop.loc[shop['도로명주소'].str.startswith('서울')]
shop_busan = shop.loc[~shop['도로명주소'].str.startswith('서울')]
```

```python
(ggplot(shop_seoul)
 + aes(x = '경도', y = '위도')
 + geom_point(color = 'gray', alpha = 0.2, size = 0.5)
 + theme(text = element_text(family = 'NanumGothic'))
)
```

```python
(ggplot(shop_busan)
 + aes(x = '경도', y = '위도')
 + geom_point(color = 'black', alpha = 0.2, size = 0.5)
 + theme(text = element_text(family = 'NanumGothic'))
)
```

```python
seoul_edu_computer = seoul_edu[seoul_edu['상권업종중분류명'] == '학원-컴퓨터']
seoul_edu_computer.shape
```

```python
import folium

geo_df = seoul_edu_computer

map = folium.Map(location=[geo_df['위도'].mean(),
                           geo_df['경도'].mean()],
                 zoom_start = 16, tiles = 'Stamen Terrain')

for n in geo_df.index:
    shop_name = geo_df.loc[n, '상호명'] + ' - ' + geo_df.loc[n, '도로명주소']
    folium.Marker([geo_df.loc[n, '위도'], geo_df.loc[n, '경도']],
                  popup = shop_name).add_to(map)
  
map
```
